```python
import numpy as np
import scipy
from scipy import linalg

# a. Define a matrix A
A = np.arange(1, 10).reshape(3, 3)

# b. Define a vector b
b = np.arange(1, 4)

# c. Solve the linear system of equations A x = b
x = scipy.linalg.solve(A, b)
result = np.matmul(A,x)

# d. Check that your solution is correct by plugging it into the equation
print(np.isclose(result, b))
print(result)
print(b)
```

```python
# e. Repeat steps a-d using a random 3x3 matrix B (instead of the vector b)
A = np.arange(1, 10).reshape(3, 3)
B = np.array([[1,2,3],[1,2,3],[1,2,3]])
x = scipy.linalg.solve(A, B)

print(np.dot(A, x))
print(B)
```

```python
# f. Solve the eigenvalue problem for the matrix A and print the eigenvalues and eigenvectors
solution = scipy.linalg.eig(A, B)
print(solution)
```

```python
# g. Calculate the inverse, determinant of A
print(scipy.linalg.inv(A))
```

```python
# h. Calculate the norm of A with different orders
print(scipy.linalg.norm(A, ord=1))
print(scipy.linalg.norm(A, ord=2))
```

```python
# Statistics:

import matplotlib.pyplot as plt
from scipy import stats
```

```python
fig=plt.figure(figsize=(10,12))

x = np.arange(20)
mu = 7

pmf = stats.poisson.pmf(x, mu)
cdf = stats.poisson.cdf(x, mu)
rvs = stats.poisson.rvs(mu, size=1000, random_state=0)

plt.subplot(3,1,1)
plt.margins(x=0)
plt.xlim([0, 17.5])
plt.plot(x, pmf, color='blue')
plt.title('Probability Mass Function')

plt.subplot(3,1,2)
plt.margins(x=0)
plt.xlim([0, 17.5])
plt.plot(x, cdf, color='orange')
plt.title('Cummulative Distribution Function')

plt.subplot(3,1,3)
plt.margins(x=0)
plt.xlim([0, 17.5])
plt.hist(rvs, x, color='green')
plt.title('Histogram of 1000 Random Realizations of the Variable')
```

```python
fig=plt.figure(figsize=(10,12))

x = np.arange(20)
loc = 7
scale = 1

pdf = stats.norm.pdf(x, loc, scale)
cdf = stats.norm.cdf(x, loc, scale)
rvs = stats.norm.rvs(loc, scale, size=1000, random_state=0)

plt.subplot(3,1,1)
plt.margins(x=0)
plt.xlim([2, 12])
plt.plot(x, pdf, color='blue')
plt.title('Probability Mass Function')

plt.subplot(3,1,2)
plt.margins(x=0)
plt.xlim([2, 12])
plt.plot(x, cdf, color='orange')
plt.title('Cummulative Distribution Function')

plt.subplot(3,1,3)
plt.margins(x=0)
plt.xlim([2, 12])
plt.hist(rvs, x, color='green')
plt.title('Histogram of 1000 Random Realizations of the Variable')
```
